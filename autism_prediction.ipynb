```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

### ADL 
ADL_log = LogisticRegression()    # get model from sklearn
ADL_log.fit(ADL_X_train, ADL_y_train) # fit with train data

###ADU
ADU_log = LogisticRegression()    # 
ADU_log.fit(ADU_X_train, ADU_y_train) #

###CHD
CHD_log = LogisticRegression()    #
CHD_log.fit(CHD_X_train, CHD_y_train) #

# Calculate accuracy, sensitivity, specificity for adult
def calculate (X_train, y_train, X_test, y_test):
  model = LogisticRegression() 
  model.fit(X_train, y_train)
  y_predict = model.predict(X_test)
  tn, fp, fn, tp = confusion_matrix(y_test, y_predict).ravel()
  specificity = tn / (tn+fp)
  sensitivity = tp/(tp + fn)
  print("=====")
  print("Specificity score: ", round(specificity,4))
  print("Sensitivity score: ", round(sensitivity,4))
  print("Accuracy score: ", round( accuracy_score(y_test, y_predict), 4))

print("\nAdult")
calculate(ADU_X_train, ADU_y_train, ADU_X_test, ADU_y_test)
print("\nChilren")
calculate(CHD_X_train, CHD_y_train, CHD_X_test, CHD_y_test)
print("\nAdolescent")
calculate(ADL_X_train, ADL_y_train, ADL_X_test, ADL_y_test)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import learning_curve  
from sklearn.model_selection import ShuffleSplit


def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None,
                        n_jobs=1, train_sizes=np.linspace(.1, 1.0, 5)):
    plt.figure()
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Cross-validation score")

    plt.legend(loc="best")
    return plt
```

```python
# Plot learning curve for Adult

title = r"Adult Learning Curves (Logistic Regression)"
cv = ShuffleSplit(n_splits=50, test_size=0.2, random_state=0)
estimator = LogisticRegression()   
plot_learning_curve(
    estimator, title, Adult_features_final, Adult_classes, ylim=(0.7, 1.01), cv=cv, n_jobs=4
)
plt.show()

title = r"Adult Learning Curves (SVM, RBF kernel, $\gamma=0.1$)"
cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
estimator = svm.SVC(kernel='rbf', gamma=0.1)
plot_learning_curve(
    estimator, title, Adult_features_final, Adult_classes, ylim=(0.7, 1.01), cv=cv, n_jobs=4
)
plt.show()

title = "Adult Learning Curves (Naive Bayes)"
cv = ShuffleSplit(n_splits=50, test_size=0.2, random_state=0)
estimator = MultinomialNB()
plot_learning_curve(
    estimator, title, Adult_features_final, Adult_classes, ylim=(0.7, 1.01), cv=cv, n_jobs=4
)
plt.show()

title = r"Adult Learning Curves (KNN, n=7)"
cv = ShuffleSplit(n_splits=50, test_size=0.2, random_state=0)
estimator = KNeighborsClassifier(n_neighbors=7)
plot_learning_curve(
    estimator, title, Adult_features_final, Adult_classes, ylim=(0.7, 1.01), cv=cv, n_jobs=4
)
plt.show()
```
